```python
### 27-35 data preprocessing

# Headers
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy
import seaborn as sns

df = pd.read_csv("dataset.csv")
df.head(10)
```

```python
df_27_35 = df.iloc[:,27:]
df_27_35.head(20)
```

```python
df_27_35.info()
```

```python
df_27_35['Target'].unique()
```

```python
# So there are 3 unique values in target column which we can replace by¶
## Dropout -> 0
## Enrolled -> 1
## Graduate -> 2
```

```python
df_27_35.drop(df_27_35[df_27_35["Target"]=="Enrolled"].index, inplace=True)
df_27_35['Target'] = df_27_35['Target'].map({
    'Dropout':0,
    'Graduate':1,
    'Enrolled':1
})
```

```python
df_27_35
```

```python
df_27_35.to_csv("Dataset_27_35_cleaned.csv", mode='w')
```

```python
df_27_35_scaled = df_27_35
```

```python
from sklearn.preprocessing import StandardScaler

# Select the columns to be scaled
cols_to_scale = ['Curricular units 2nd sem (grade)', 'Unemployment rate', 'Inflation rate', 'GDP']

# Create a StandardScaler object
scaler = StandardScaler()

# Fit the scaler to the selected columns
scaler.fit(df_27_35_scaled[cols_to_scale])

# Scale the selected columns
df_27_35_scaled[cols_to_scale] = scaler.transform(df_27_35_scaled[cols_to_scale])
```

```python
df_27_35_scaled
```

```python
df_27_35_scaled.to_csv("Dataset_27_35_scaled.csv", mode='w')
```
